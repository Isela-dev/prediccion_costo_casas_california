```python
import pandas as pd
import seaborn as sb

# Convertir "population" a numrico
datos["population"] = pd.to_numeric(datos["population"], errors="coerce")

# Dividir por 100 y convertir a float
sizes = datos["population"] / 100
sizes = sizes.astype(float)

# Crear el grfico de dispersin
sb.scatterplot(x="latitude", y="longitude", data=datos, hue="median_house_value", palette="coolwarm", size=sizes)
```

```python
import pandas as pd

# Crear variables dummies para la columna "ocean_proximity" del DataFrame "datos_na"
dummies = pd.get_dummies(datos_na["ocean_proximity"], dtype=int)

# Unir las variables dummies al DataFrame "datos_na"
datos_na = datos_na.join(dummies)

# Mostrar las primeras filas del DataFrame con las variables dummies
print(datos_na.head())
```
